```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')

titanic.head()
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
print(titanic['sex'].value_counts())
print(titanic['class'].value_counts())
print(titanic['survived'].value_counts())
```

```python
sns.pairplot(titanic[['age', 'fare', 'pclass', 'survived']], hue='survived')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
titanic['age'].hist(bins=30)
plt.title('Age Distribution')
plt.show()
```

```python
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Age Distribution by Passenger Class')
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic)
plt.title('Age vs Fare Colored by Survival')
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title('Survival Count by Gender')
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=titanic)
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x='age', hue='survived', bins=30, multiple='stack')
plt.title('Age Distribution by Survival Status')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.ylim(0, 300)  # Limit y-axis to see majority of data
plt.title('Fare Distribution by Passenger Class')
plt.show()
```
